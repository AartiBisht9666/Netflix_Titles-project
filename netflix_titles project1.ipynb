```python
import numpy as np

# =====================================
# Missing Value Treatment
# =====================================

# 1. Director Column
df['director'] = df['director'].fillna('Unknown')

# 2. Cast Column
df['cast'] = df['cast'].fillna('Not Available')

# 3. Country Column
# Director-wise mode imputation
df['country'] = df.groupby('director')['country'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# Remaining missing countries
df['country'] = df['country'].fillna('Unknown')

# 4. Duration Column
# Type-wise mode imputation (improved version)
df['duration'] = df.groupby('type')['duration'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# =====================================
# Final Check
# =====================================

print(df.isnull().sum())
```

## Missing Value Treatment

Before conducting the exploratory data analysis, a detailed assessment of missing values was performed to improve data quality
and ensure reliable insights. The dataset contained missing values in the `director`, `cast`, `country`, `date_added`, `rating`,
and `duration` columns.

To handle missing values appropriately, different imputation strategies were applied based on the nature and importance of each
feature.

### Director

The `director` column contained a large number of missing values. Since director information is categorical and highly diverse,
missing values were replaced with the label **"Unknown"** to preserve records without introducing incorrect assumptions.

### Cast

Missing values in the `cast` column were replaced with **"Not Available"**. This approach retained all observations while clearly
indicating unavailable cast information.

### Country

For the `country` column, a director-based imputation strategy was applied. Missing country values were filled using the most 
frequent country associated with the same director. Any remaining missing values were replaced with **"Unknown"**. This method
provided more context-aware imputation than using a global mode.

### Duration

The `duration` column contained only a few missing observations. Missing values were imputed using the most frequent duration
within each content type (`Movie` or `TV Show`). This ensured consistency between different categories of content.

### Date Added and Rating

The `date_added` and `rating` columns contained only a negligible number of missing values. To preserve data authenticity and avoid
introducing artificial information, these missing values were retained and excluded only when required for specific analyses.

### Outcome

After preprocessing, the major categorical attributes were successfully cleaned while maintaining data integrity.
The remaining missing values represented less than 1% of the dataset and were considered insignificant for the overall
exploratory analysis.


```python
# Convert date column
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 1. Year Added
df['year_added'] = df['date_added'].dt.year

# 2. Content Age
df['content_age'] = df['year_added'] - df['release_year']

# 3. Main Genre
df['main_genre'] = df['listed_in'].astype(str).apply(
    lambda x: x.split(',')[0].strip()
)

# 4. Primary Country
df['primary_country'] = df['country'].astype(str).apply(
    lambda x: x.split(',')[0].strip()
)

# 5. Duration Numeric
df['duration_num'] = pd.to_numeric(
    df['duration'].astype(str).str.extract(r'(\d+)')[0],
    errors='coerce'
)

# 6. Genre Count
df['genre_count'] = df['listed_in'].astype(str).apply(
    lambda x: len(x.split(','))
)

# Check
print(df[['year_added',
          'content_age',
          'main_genre',
          'primary_country',
          'duration_num',
          'genre_count']].head())
```

## Feature Engineering

To enhance the analytical value of the dataset, a set of meaningful features was created from existing attributes.
The `year_added` feature was extracted from the `date_added` column to analyze Netflix content additions over time.
A new feature named `content_age` was calculated as the difference between the year a title was added to Netflix and
its original release year, helping measure content recency.

The `main_genre` feature was derived from the first listed category in the `listed_in` column to identify the primary
genre of each title. Similarly, `primary_country` was extracted from the `country` column to represent the main country 
of origin. The `duration_num` feature was created by extracting the numeric component from the duration field, enabling
quantitative duration analysis. Finally, `genre_count` was generated to measure the number of genres associated with each
title, providing insights into genre diversity.

These engineered features improved the dataset's analytical capability and supported deeper exploratory analysis.
